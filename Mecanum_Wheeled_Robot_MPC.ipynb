# Mecanum Wheeled Robot:

A mecanum wheeled robot, gets its name from Mecanum AB company, is a omnidirectional wheeled robot designed for land-based 2-D vehicle motion to move in any direction.

# Mechanisms of Movement of Mecanum Wheeled Robot:

1. When all four wheels move in the same direction at the same speed, the vehicle moves forward or backward due to longitudinal forces combining while transverse forces cancel.

2. When one side's wheels move in one direction while the other side's wheels move in the opposite direction at the same speed, the vehicle rotates in place. Transverse forces cancel, but longitudinal forces generate a torque.

3. If diagonal wheels move in one direction while the other diagonal wheels move in the opposite direction at the same speed, the vehicle moves sideways. Transverse forces add up, while longitudinal forces cancel.
